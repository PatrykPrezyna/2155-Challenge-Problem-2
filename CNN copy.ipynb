```python
# Load data and run a small hyperparameter sweep for the CNN model
# By default this runs for advisor 0 only to keep runtime reasonable — set advisors = range(4) to run all advisors.

grids = load_grids()
ratings = np.load("datasets/scores.npy")
ratings_df = pd.DataFrame(ratings, columns=["Wellness", "Tax", "Transportation", "Business"])

advisors = [0]  # change to list(range(4)) to train for all advisors
all_predictions = []
hyper_results = {}  # store results per advisor

for advisor in advisors:
    print(f"\nRunning hyperparameter sweep for Advisor {advisor}")
    grids_subset, ratings_subset = select_rated_subset(grids, ratings[:, advisor])
    X_train, X_validation, Y_train, Y_validation = train_test_split(
        grids_subset, ratings_subset, test_size=0.2, random_state=42
    )

    height, width = 7, 7
    num_pixel_classes = 5
    X_train_encoded = tf.one_hot(X_train, depth=num_pixel_classes)
    X_validation_encoded = tf.one_hot(X_validation, depth=num_pixel_classes)
    grids_encoded = tf.one_hot(grids, depth=num_pixel_classes)

    # Model factory using parameter dictionary
    def create_model(params):
        model = models.Sequential()
        model.add(layers.Input(shape=(height, width, num_pixel_classes)))

        # Add convolutional blocks according to filters list
        for i, f in enumerate(params['filters']):
            model.add(layers.Conv2D(f, kernel_size=params.get('kernel_size', 3), activation='relu', padding='same'))
            model.add(layers.BatchNormalization())
            # second conv in block (optional)
            if params.get('double_conv', True):
                model.add(layers.Conv2D(f, kernel_size=params.get('kernel_size', 3), activation='relu', padding='same'))
                model.add(layers.BatchNormalization())
            model.add(layers.MaxPooling2D(pool_size=(2, 2)))
            if params.get('dropout_rate', 0) > 0:
                model.add(layers.Dropout(params['dropout_rate']))

        model.add(layers.Flatten())
        model.add(layers.Dense(params.get('dense_units', 128), activation='relu'))
        model.add(layers.BatchNormalization())
        if params.get('dense_dropout', 0) > 0:
            model.add(layers.Dropout(params['dense_dropout']))
        model.add(layers.Dense(1, activation='linear'))

        optimizer = tf.keras.optimizers.Adam(learning_rate=params.get('learning_rate', 1e-3))
        model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])
        return model

    # Small hyperparameter grid (easy to extend)
    hyperparameter_list = [
        {'name': 'base', 'filters': [32, 64], 'kernel_size': 3, 'dropout_rate': 0.0, 'dense_units': 128, 'dense_dropout': 0.0, 'learning_rate': 1e-3},
        {'name': 'dropout', 'filters': [32, 64], 'kernel_size': 3, 'dropout_rate': 0.25, 'dense_units': 128, 'dense_dropout': 0.5, 'learning_rate': 1e-3},
        {'name': 'more_filters', 'filters': [64, 128], 'kernel_size': 3, 'dropout_rate': 0.25, 'dense_units': 256, 'dense_dropout': 0.5, 'learning_rate': 5e-4},
        {'name': 'small_lr', 'filters': [32, 64], 'kernel_size': 3, 'dropout_rate': 0.25, 'dense_units': 128, 'dense_dropout': 0.5, 'learning_rate': 1e-4}
    ]

    advisor_results = []

    for params in hyperparameter_list:
        print(f"\nTraining config: {params['name']} (lr={params['learning_rate']}, filters={params['filters']}, dropout={params['dropout_rate']})")
        model = create_model(params)

        callbacks = [
            tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True),
            tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6)
        ]

        history = model.fit(
            X_train_encoded, Y_train,
            validation_data=(X_validation_encoded, Y_validation),
            epochs=50,
            batch_size=32,
            callbacks=callbacks,
            verbose=1
        )

        # Evaluate final validation metrics
        val_loss = history.history['val_loss'][-1]
        val_mae = history.history['val_mae'][-1]

        advisor_results.append({'params': params, 'model': model, 'history': history, 'val_loss': val_loss, 'val_mae': val_mae})

    # Store results
    hyper_results[advisor] = advisor_results

    # Pick best model by validation MAE
    best = min(advisor_results, key=lambda r: r['val_mae'])
    print(f"\nBest config for advisor {advisor}: {best['params']['name']} with val_mae={best['val_mae']:.4f}")

    # Produce predictions using best model and merge with known ratings
    predictions = best['model'].predict(grids_encoded)
    mask = np.where(~np.isnan(ratings[:, advisor]))
    predictions[mask, 0] = ratings[:, advisor][mask]
    all_predictions.append(predictions)

# After sweep: visualize comparisons across hyperparameters for the (first) advisor
import pandas as _pd

for advisor in hyper_results:
    results = hyper_results[advisor]
    # Plot validation MAE and loss for each config
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    for r in results:
        label = r['params']['name']
        plt.plot(r['history'].history['val_mae'], label=f"{label} (val_mae final {r['val_mae']:.4f})")
    plt.title(f'Validation MAE per epoch (Advisor {advisor})')
    plt.xlabel('Epoch')
    plt.ylabel('Validation MAE')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    for r in results:
        label = r['params']['name']
        plt.plot(r['history'].history['val_loss'], label=f"{label} (val_loss final {r['val_loss']:.4f})")
    plt.title(f'Validation Loss (MSE) per epoch (Advisor {advisor})')
    plt.xlabel('Epoch')
    plt.ylabel('Validation Loss (MSE)')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Summary table
    summary = _pd.DataFrame([{
        'config': r['params']['name'],
        'val_mae': r['val_mae'],
        'val_loss': r['val_loss'],
        'filters': str(r['params']['filters']),
        'dropout': r['params']['dropout_rate'],
        'lr': r['params']['learning_rate']
    } for r in results])

    display(summary.sort_values('val_mae'))

# Stack predictions from best models (one per advisor) and continue with selection logic as before
all_predictions = [np.squeeze(a) for a in all_predictions]
final_prediction_array = np.stack(all_predictions).T
min_predictions = np.min(final_prediction_array, axis=1)
print(f"Number of valid grids (as predicted): {np.sum(min_predictions > 0.75)}")
top_100_indices = np.argpartition(min_predictions, -100)[-100:]
top_100_grids = grids[top_100_indices]
```
